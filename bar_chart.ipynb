```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [30, 60, 45, 80, 20]

# Condition ke hisab se plot karte hain
for i in range(len(x)):
    if y[i] > 50:
        plt.scatter(x[i], y[i], color='green', label='> 50' if i == 0 else "")
    else:
        plt.scatter(x[i], y[i], color='red', label='< 50' if i == 0 else "")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Conditional Legend Example")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

player = ['Sachin','Rohit','Virat','Rahul','Dhoni','Shikhar','Surya']
matches = [500,450,470,200,430,350,100]

# Bar color decide karte hain
# colors = ['green' if m > 400 else 'red' for m in matches]

# Bar plot
for i in range(len(player)):
    if matches[i] > 400:
        plt.bar(player[i], matches[i], color='green', label='> 400 Matches' if i == 0 else "")
    else:
        plt.bar(player[i], matches[i], color='red', label=' 400 Matches' if i == 0 else "")

# Labels aur title
plt.xlabel("Players")
plt.ylabel("Matches Played")
plt.title("Players Match Count (Conditional Legend)")

# Legend show karo
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
player = ['Sachin','Rohit','Virat','Rahul','Dhoni','Shikhar','Surya']
matches = [500,450,470,200,430,350,100]

# DataFrame banao
df = pd.DataFrame({
    'Player': player,
    'Matches': matches
})

# Condition ke hisab se ek naya column add karo
df['Category'] = df['Matches'].apply(lambda x: '> 400 Matches' if x > 400 else ' 400 Matches')

# Seaborn plot
sns.barplot(
    data=df,
    x='Player',
    y='Matches',
    hue='Category',       # legend yahan se banega automatically
    palette={'> 400 Matches':'green', ' 400 Matches':'red'}
)

plt.title("Players Match Count (Seaborn Conditional Legend)")
plt.xlabel("Players")
plt.ylabel("Matches Played")
plt.show()
```
