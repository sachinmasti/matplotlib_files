# Pie Chart using Matplotlib
**English:** This notebook demonstrates how to create various types of pie charts using the Matplotlib library in Python.
**Hindi:** Yeh notebook Python mein Matplotlib library ka upyog karke alag-alag prakaar ke pie chart banane ka tarika dikhata hai.

**English:** Importing the `matplotlib.pyplot` module, which is essential for plotting.
**Hindi:** `matplotlib.pyplot` module ko import karna, jo plotting ke liye zaroori hai.

```python
import matplotlib.pyplot as plt
```

**English:** Setting the visual style of the plots to `'ggplot'`. This style is popular for its visually appealing aesthetics.
**Hindi:** Plots ke visual style ko `'ggplot'` par set karna. Yeh style apne sundar visuals ke liye lokpriya hai.

```python
plt.style.use('ggplot')
```

**English:** This cell defines the data (labels and runs) for a simple pie chart and then generates the plot to show the career runs of different Indian batsmen.
**Hindi:** Yeh cell ek sadharan pie chart ke liye data (labels aur runs) define karta hai aur phir alag-alag भारतीय बल्लेबाजों ke career runs dikhane ke liye plot banata hai.

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
plt.show()
```

**English:** Defining new data for a different set of players.
**Hindi:** Alag-alag khiladiyon ke ek naye set ke liye naya data define karna.

```python
labels = ['sachin','rohit','virat','rahul','dhoni']
runs = [1232,1221,2132,3221,2123]
```

**English:** Creating a more advanced and styled pie chart. This includes:
- `autopct`: To display the percentage value on each slice.
- `colors`: To set custom colors for the slices.
- `wedgeprops`: To define properties of the wedges, like a black edge.
- `shadow`: To add a shadow effect to the pie chart.
**Hindi:** Ek adhik advanced aur styled pie chart banana. Ismein shamil hai:
- `autopct`: Har slice par pratishat value dikhane ke liye.
- `colors`: Slices ke liye custom rang set karne ke liye.
- `wedgeprops`: Wedges ke gun-dharmo ko paribhashit karne ke liye, jaise ki kala kinara.
- `shadow`: Pie chart mein shadow prabhav jodne ke liye.

```python
explod = [0,0.1,0,0,0]
colors = ['#1E4928','#B1CBDA','#A69CC7','#DDA8E3']
plt.pie(runs,labels=labels , autopct='%1.1f%%',colors = colors,wedgeprops={'edgecolor':'black','linestyle':"--","linewidth":2},shadow=True)
plt.title('compare with pie chart')
```

**English:** This cell demonstrates the correct way to create a donut chart. It's done by using the `wedgeprops` argument and setting a `width` for the wedges. This creates a hole in the center.
**Hindi:** Yeh cell ek donut chart banane ka sahi tarika dikhata hai. Yeh `wedgeprops` argument ka upyog karke aur wedges ke liye ek `width` set karke kiya jata hai. Isse center mein ek chhed ban jata hai.

```python
plt.pie(runs,labels=labels,colors=['#C4B3C5','#D3E6EB','#CC9BCB','#CDAAB6'],explode=explod,wedgeprops=dict(width = 0.2,edgecolor ='w'),startangle=100)
plt.show()
```
