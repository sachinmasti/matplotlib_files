### Seaborn vs. Matplotlib: What's the Difference?

**English:**

*   **Matplotlib:** Think of Matplotlib as the foundation. It's a powerful low-level library that gives you complete control over every detail of your plot. This is great for creating highly customized, publication-quality graphics, but it can require more code to make things look polished.
*   **Seaborn:** Seaborn is built on top of Matplotlib. Its main goal is to make creating beautiful and statistically informative plots easier. It comes with excellent default styles and color palettes and has simple functions for creating complex plots like violin plots and heatmaps, which would be much harder to make in Matplotlib alone.

**Hindi:**

*   **Matplotlib:** Matplotlib ko neev samjhein. Yeh ek shaktishali low-level library hai jo aapko apne plot ke har detail par poora control deti hai. Yeh highly customized, publication-quality graphics banane ke liye accha hai, lekin cheezon ko polished dikhane ke liye aksar adhik code ki zaroorat hoti hai.
*   **Seaborn:** Seaborn Matplotlib ke upar banaya gaya hai. Iska mukhya uddeshya sundar aur statistical plots ko aasan banana hai. Ismein behtareen default styles aur color palettes hote hain aur ismein violin plots aur heatmaps jaise jatil plots banane ke liye saral functions hote hain, jo akele Matplotlib mein banana bahut mushkil hoga.

*   **Use Seaborn for:** Quick data exploration, creating attractive statistical visualizations, and working with pandas DataFrames.

**English:** Importing the `matplotlib.pyplot` module for plotting.

**Hindi:** Plotting ke liye `matplotlib.pyplot` module ko import karna.

```python
import matplotlib.pyplot as plt
```

**English:** Defining the data for the bar chart: a list of players and their corresponding number of matches.

**Hindi:** Bar chart ke liye data define karna: khiladiyon ki ek list aur unke corresponding matches ki sankhya.

```python
player = ['Sachin','Rohit','Virat','Rahul','Dhoni','Shikhar','Surya']
matches = [500,450,470,200,430,350,100]
```

**English:** Combining the player and matches lists into a single list. This isn't used in the subsequent plots but demonstrates how to group related data.

**Hindi:** Player aur matches lists ko ek hi list mein combine karna. Yeh aage ke plots mein upyog nahi kiya gaya hai, lekin yah dikhata hai ki related data ko kaise group kiya jaata hai.

```python
combine = [player,matches]
```

**English:** Creating a list of colors. A list comprehension is used to set the color of the bar to a light green if the number of matches is less than 300, and a dark grey otherwise.

**Hindi:** Rango ki ek list banana. Ek list comprehension ka upyog karke bar ke rang ko halka hara set kiya gaya hai agar matches ki sankhya 300 se kam hai, aur anyatha gehra grey.

```python
color = ['#B9EFA3' if i < 300 else'#686277' for i in matches]
```

**English:** Creating the bar chart using the original logic. The `color` list sets the color of each bar individually. A legend is attempted, but since the bars themselves don't have individual labels, it won't display anything.

**Hindi:** Mool tark ka upyog karke bar chart banana. `color` list har bar ke rang ko alag se set karti hai. Ek legend banane ka prayas kiya gaya hai, lekin chunki bars ke paas alag-alag labels nahi hain, isliye yah kuchh bhi pradarshit nahi karega.

```python
plt.bar(player,matches,color=color)
plt.title('payers machetes')
plt.xlabel('players')
plt.ylabel('matches')
plt.legend()
plt.tight_layout()
```

**English:** This is a more robust way to create a bar chart with a conditional legend. It iterates through the players and matches, plotting each bar one by one. This allows us to assign a specific label (' > 400 Matches' or '≤ 400 Matches') to the first bar of each color group, ensuring the legend displays correctly.

**Hindi:** Yeh conditional legend ke saath ek bar chart banane ka ek adhik majboot tareeka hai. Yeh khiladiyon aur matches ke madhyam se iterate karta hai, har bar ko ek-ek karke plot karta hai. Isse humein har rang samuh ke pehle bar ko ek vishisht label (' > 400 Matches' ya '≤ 400 Matches') assign karne ki anumati milti hai, jisse yah sunishchit hota hai ki legend sahi dhang se pradarshit ho.

```python
for i in range(len(player)):
    if matches[i] > 400:
        plt.bar(player[i], matches[i],color='green',label='mathes > 400' if i == 0 else "")
    else:
        plt.bar(player[i], matches[i], color='red', label='≤ 400 Matches' if i == 0 else "")

# Labels aur title
plt.xlabel("Players")
plt.ylabel("Matches Played")
plt.title("Players Match Count (Conditional Legend)")

# Legend show karo
plt.legend()
plt.show()
```
