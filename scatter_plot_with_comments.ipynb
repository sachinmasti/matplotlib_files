# Scatter Plot using Matplotlib

**English:** This notebook demonstrates how to create scatter plots using the Matplotlib library.

**Hindi:** Yeh notebook Matplotlib library ka upyog karke scatter plot banane ka tarika dikhata hai.

**English:** Importing the `matplotlib.pyplot` module for plotting.

**Hindi:** Plotting ke liye `matplotlib.pyplot` module ko import karna.

```python
import matplotlib.pyplot as plt
```

**English:** Defining two lists, `hours` and `marks`, which will serve as the x and y coordinates for our scatter plot.

**Hindi:** Do lists, `hours` aur `marks`, define karna, jo hamare scatter plot ke liye x aur y coordinates ka kaam karenge.

```python
hours = [2,3,4,2,2,3,4,3,2,2]
marks = [56,78,54,87,54,87,78,24,53,76]
```

**English:** This cell creates a scatter plot where the color of each point (`c`) is determined by the `sizes` list (derived from marks), using the 'plasma' colormap. A color bar is added to show the mapping of colors to marks.

**Hindi:** Yeh cell ek scatter plot banata hai jahan har point ka rang (`c`) `sizes` list (jo marks se derive kiya gaya hai) dwara tay hota hai, 'plasma' colormap ka upyog karke. Rango ki marks se mapping dikhane ke liye ek color bar joda gaya hai.

**For** a `cmap` view this *link*
>https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
sizes = [s * 2 for s in marks]
plt.scatter(hours,marks ,c =sizes, cmap='plasma')
plt.colorbar(label='Marks')
plt.grid(True)
plt.title('marks by study hours')
```

**English:** This cell shows how to plot multiple datasets on a single scatter plot. We define data for two classes, 'Class A' and 'Class B', and plot them with different colors and labels for easy comparison.

**Hindi:** Yeh cell ek hi scatter plot par multiple datasets ko kaise plot karein, yeh dikhata hai. Hum do classes, 'Class A' aur 'Class B' ke liye data define karte hain aur unhe aasaan tulna ke liye alag-alag rangon aur labels ke saath plot karte hain.

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
