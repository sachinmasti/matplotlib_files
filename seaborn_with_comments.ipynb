**English:** This notebook provides a brief introduction to Seaborn and highlights the differences between Seaborn and Matplotlib.

**Hindi:** Yeh notebook Seaborn ka ek chhota sa parichay deta hai aur Seaborn aur Matplotlib ke beech ke antar ko highlight karta hai.

### What is the difference between Seaborn and Matplotlib?

**English:**

*   **Matplotlib** is a powerful and flexible low-level library that gives you full control over your plots. You can customize every single element of a chart, but it often requires writing more code to get a visually appealing result.
*   **Seaborn** is a high-level library built on top of Matplotlib. It's designed to make creating beautiful and statistically informative plots much easier and faster. It comes with built-in themes and color palettes, and it simplifies plotting complex statistical data.

**Hindi:**

*   **Matplotlib** ek powerful aur flexible low-level library hai jo aapko apne plots par poora control deti hai. Aap ek chart ke har ek element ko customize kar sakte hain, lekin ek accha dikhne wala result paane ke liye aksar zyada code likhna padta hai.
*   **Seaborn** Matplotlib ke upar bani ek high-level library hai. Iska design sundar aur statistical plots ko bahut aasaan aur tezi se banane ke liye kiya gaya hai. Ismein built-in themes aur color palettes hote hain, aur yeh complex statistical data ko plot karna aasaan banata hai.

### Use Cases:

**English:**

*   **Use Matplotlib when:** You need detailed, custom plots for publications or presentations and want fine-grained control over every aspect.
*   **Use Seaborn when:** You want to quickly explore your data, create aesthetically pleasing statistical plots (like heatmaps, violin plots, etc.), and write less code.

**Hindi:**

*   **Matplotlib ka upyog karein jab:** Aapko publications ya presentations ke liye detailed, custom plots ki zaroorat ho aur har aspect par fine-grained control chahiye.
*   **Seaborn ka upyog karein jab:** Aap apne data ko jaldi se explore karna chahte hain, sundar statistical plots (jaise heatmaps, violin plots, etc.) banana chahte hain, aur kam code likhna chahte hain.

**English:** Importing the `seaborn` and `numpy` libraries. Seaborn is used for statistical data visualization, and NumPy is used for numerical operations.

**Hindi:** `seaborn` aur `numpy` libraries ko import karna. Seaborn statistical data visualization ke liye aur NumPy numerical operations ke liye upyog hota hai.

```python
import seaborn as sns
import numpy as np
```

**English:** Creating a NumPy array of numbers that will be used as our x-axis data.

**Hindi:** Ek NumPy array banana jo hamare x-axis data ke roop mein upyog hoga.

```python
x = np.array([12,3,32,34,14,56,67,67,54,34,93,23,3,234,43,84,56,56,67,7,67,65,54,43,4,43])
```

**English:** Setting the visual theme for all subsequent plots to 'darkgrid' using Seaborn. This automatically adds a dark grid and styles the plot.

**Hindi:** Seaborn ka upyog karke aage ke sabhi plots ke liye visual theme ko 'darkgrid' par set karna. Yeh apne aap ek dark grid jod deta hai aur plot ko style karta hai.

```python
sns.set_theme(style="darkgrid")
```

**English:** Calculating the sine of each number in the `x` array to create our y-axis data.

**Hindi:** Y-axis data banane ke liye `x` array mein har sankhya ka sine calculate karna.

```python
y = np.sin(x)
```

**English:** Creating a line plot of the `x` and `y` data using Seaborn's `lineplot` function. The 'darkgrid' theme we set earlier is automatically applied.

**Hindi:** Seaborn ke `lineplot` function ka upyog karke `x` aur `y` data ka ek line plot banana. Humne jo 'darkgrid' theme pehle set kiya tha, woh apne aap apply ho jaata hai.

```python
sns.lineplot(x = x , y= y)
```
