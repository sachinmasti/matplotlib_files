# Subplot using Matplotlib

**English:** This notebook demonstrates how to create multiple subplots in a single figure using Matplotlib.

**Hindi:** Yeh notebook Matplotlib ka upyog karke ek hi figure mein multiple subplots banane ka tarika dikhata hai.

**English:** Importing the `matplotlib.pyplot` module, which is essential for any plotting tasks.

**Hindi:** Plotting ke kisi bhi kaam ke liye zaroori `matplotlib.pyplot` module ko import karna.

```python
import matplotlib.pyplot as plt
```

**English:** Defining the data. `nums` is a list of numbers, and `x` and `y` are lists containing the square and cube of those numbers, respectively.

**Hindi:** Data define karna. `nums` sankhyaon ki ek list hai, aur `x` aur `y` un sankhyaon ke square aur cube wali lists hain.

```python
nums = list(range(1,100))
x = [i**2 for i in nums]
y = [i**3 for i in nums]
```

**English:** This cell creates a figure with two subplots side-by-side (1 row, 2 columns). The first subplot plots `nums` vs `x` (squares), and the second plots `nums` vs `y` (cubes). `tight_layout()` adjusts the plots to prevent them from overlapping.

**Hindi:** Yeh cell ek figure banata hai jisme do subplots ek saath (1 row, 2 columns) hain. Pehla subplot `nums` vs `x` (squares) plot karta hai, aur doosra `nums` vs `y` (cubes) plot karta hai. `tight_layout()` plots ko overlap hone se bachane ke liye adjust karta hai.

```python
plt.subplot(1,2,1,)
plt.plot(nums,x)
plt.title('performance using squre')
plt.xlabel('numbers')
plt.ylabel('square of numbers')

plt.subplot(1,2,2)
plt.plot(nums,y)
plt.title('performance using cube')
plt.xlabel('number')
plt.ylabel('cube of numbers')
plt.tight_layout()
```

**English:** This cell shows a more object-oriented approach to creating subplots. `plt.subplots()` returns a figure and an array of axes. We then plot on each axis individually. This method offers more control over the figure and plots.

**Hindi:** Yeh cell subplots banane ka ek adhik object-oriented approach dikhata hai. `plt.subplots()` ek figure aur axes ka ek array return karta hai. Phir hum har ek axis par alag se plot karte hain. Yeh method figure aur plots par adhik control deta hai.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(nums, x)
axs[0].set_title('x ** 2')

axs[1].plot(nums, y)
axs[1].set_title('y cube')
 
fig.suptitle('Simple Comparison Plots', fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.85)  # So title doesn't overlap
fig.savefig('my_plots.png')    # Save as image

plt.show()
```
