# Plotting using Seaborn

**English:** This notebook demonstrates various plotting techniques using the Seaborn library, which is built on top of Matplotlib.

**Hindi:** Yeh notebook Seaborn library ka upyog karke alag-alag plotting techniques dikhata hai, jo Matplotlib ke upar bani hai.

### What is the difference between Seaborn and Matplotlib?

**English:**

*   **Matplotlib** is a powerful and flexible low-level library that gives you full control over your plots. You can customize every single element of a chart, but it often requires writing more code to get a visually appealing result.
*   **Seaborn** is a high-level library built on top of Matplotlib. It's designed to make creating beautiful and statistically informative plots much easier and faster. It comes with built-in themes and color palettes, and it simplifies plotting complex statistical data.

**Hindi:**

*   **Matplotlib** ek powerful aur flexible low-level library hai jo aapko apne plots par poora control deti hai. Aap ek chart ke har ek element ko customize kar sakte hain, lekin ek accha dikhne wala result paane ke liye aksar zyada code likhna padta hai.
*   **Seaborn** Matplotlib ke upar bani ek high-level library hai. Iska design sundar aur statistical plots ko bahut aasaan aur tezi se banane ke liye kiya gaya hai. Ismein built-in themes aur color palettes hote hain, aur yeh complex statistical data ko plot karna aasaan banata hai.

### Use Cases:

**English:**

*   **Use Matplotlib when:** You need detailed, custom plots for publications or presentations and want fine-grained control over every aspect.
*   **Use Seaborn when:** You want to quickly explore your data, create aesthetically pleasing statistical plots (like heatmaps, violin plots, etc.), and write less code.

**Hindi:**

*   **Matplotlib ka upyog karein jab:** Aapko publications ya presentations ke liye detailed, custom plots ki zaroorat ho aur har aspect par fine-grained control chahiye.
*   **Seaborn ka upyog karein jab:** Aap apne data ko jaldi se explore karna chahte hain, sundar statistical plots (jaise heatmaps, violin plots, etc.) banana chahte hain, aur kam code likhna chahte hain.

**English:** Importing the `seaborn` and `matplotlib.pyplot` libraries. Seaborn is used for high-level plotting, and Matplotlib is often used alongside it for customization.

**Hindi:** `seaborn` aur `matplotlib.pyplot` libraries ko import karna. Seaborn high-level plotting ke liye aur Matplotlib aksar iske saath customization ke liye upyog hota hai.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**English:** Printing the names of all the built-in datasets available in Seaborn.

**Hindi:** Seaborn mein uplabdh sabhi built-in datasets ke naam print karna.

```python
print(sns.get_dataset_names())
```

**English:** Loading the 'flights' dataset from Seaborn into a pandas DataFrame.

**Hindi:** Seaborn se 'flights' dataset ko ek pandas DataFrame mein load karna.

```python
flights = sns.load_dataset('flights')
```

**English:** Displaying the first few rows of the 'flights' DataFrame to get an idea of its structure and data.

**Hindi:** 'flights' DataFrame ke pehle kuch rows ko display karna, jisse uske structure aur data ka andaza ho.

```python
flights.head()
```

**English:** Getting information about the 'flights' DataFrame, including the data types of each column and the number of non-null values.

**Hindi:** 'flights' DataFrame ke baare mein jaankari prapt karna, jisme har column ke data types aur non-null values ki sankhya shamil hai.

```python
flights.info()
```

**English:** Setting the visual theme of the plots to 'whitegrid', which provides a clean look with a white grid.

**Hindi:** Plots ke visual theme ko 'whitegrid' par set karna, jo ek saaf look deta hai jisme safed grid hota hai.

```python
sns.set_theme(style='whitegrid')
```

**English:** Creating a line plot to visualize the trend of passenger numbers over the years.

**Hindi:** Saal dar saal passenger numbers ke trend ko visualize karne ke liye ek line plot banana.

```python
sns.lineplot(x='year',y='passengers',data=flights)
plt.title('passenger growth by year on year',c='#083207')
plt.tight_layout()
```

**English:** Creating a bar plot to show the number of passengers for each year.

**Hindi:** Har saal ke liye passengers ki sankhya dikhane ke liye ek bar plot banana.

```python
sns.barplot(x='year',y='passengers',data=flights)
plt.title('passenger growth by year on year',c='#083207')
plt.tight_layout()
```

**English:** A violin plot is a combination of a box plot and a kernel density estimate. It shows the distribution of passenger numbers for each year.

**Hindi:** Ek violin plot ek box plot aur ek kernel density estimate ka combination hai. Yeh har saal ke liye passenger numbers ke distribution ko dikhata hai.

```python
sns.violinplot(x='year',y='passengers',data=flights)
plt.title('passenger growth by year on year',c='#083207')
plt.tight_layout()
```

**English:** Displaying the full 'flights' DataFrame.

**Hindi:** Poora 'flights' DataFrame display karna.

```python
flights
```

**English:** Creating a pivot table from the 'flights' DataFrame, with years as the index, months as columns, and passenger numbers as values.

**Hindi:** 'flights' DataFrame se ek pivot table banana, jisme index ke roop mein saal, columns ke roop mein mahine, aur values ke roop mein passenger numbers hain.

```python
pv = flights.pivot(index = 'year',columns='month',values= 'passengers')
```

**English:** Creating a heatmap from the pivot table. The `annot=True` argument displays the passenger numbers on the heatmap, and `fmt='d'` formats them as integers.

**Hindi:** Pivot table se ek heatmap banana. `annot=True` argument heatmap par passenger numbers ko display karta hai, aur `fmt='d'` unhe integers ke roop mein format karta hai.

```python
sns.heatmap(pv,annot=True,fmt='d',cmap = 'magma')
```
