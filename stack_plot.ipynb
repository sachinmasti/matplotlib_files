# Stack Plot using Matplotlib
**English:** This notebook explains how to create stack plots using the Matplotlib library.
**Hindi:** Yeh notebook Matplotlib library ka upyog karke stack plot banane ka tarika batata hai.

**English:** Importing the `matplotlib.pyplot` module for plotting.
**Hindi:** Plotting ke liye `matplotlib.pyplot` module ko import karna.

```python
import matplotlib.pyplot as plt
```

**English:** Defining the initial data for a pie chart, showing hours spent on different activities.
**Hindi:** Pie chart ke liye shuruaati data define karna, jo alag-alag gatividhiyon mein bitaye gaye ghanton ko dikhata hai.

```python
activity = ['study','sleep','netfix','running']
hours = [4,6,2,1]
```

**English:** Creating a pie chart to visualize the distribution of daily activities. The `figsize` is set, colors and labels are provided, and `autopct` displays the percentage for each slice.
**Hindi:** Dainik gatividhiyon ke vitaran ko visualize karne ke liye ek pie chart banana. `figsize` set kiya gaya hai, rang aur labels diye gaye hain, aur `autopct` har slice ke liye pratishat dikhata hai.

```python
plt.figure(figsize=(6,6))
plt.pie(hours,labels=activity,colors=['green','pink','brown','red'],autopct='%1.1f%%')
plt.title('student activity')
plt.axis('equal')
```

**English:** Defining new data for a stack plot. This data represents hours spent on different activities over a week.
**Hindi:** Stack plot ke liye naya data define karna. Yeh data ek hafte mein alag-alag gatividhiyon mein bitaye gaye ghanton ko represent karta hai.

```python
days = list(range(1,8))
stydy = [3,4,2,6,1,5,2]
tv = [3,4,2,2,1,4,1]
playing = [3,4,3,5,1,2,3]
gym = [1,2,1,1,2,3,1]
```

**English:** Defining labels and colors for the stack plot.
**Hindi:** Stack plot ke liye labels aur rang define karna.

```python
label = ['days','study','tv','playing','gym']
colors  = ['#826D8F','#B7EDE6','#1E423D','#EACDAC','#D1D59C','#EAB6DB','#175216']
```

**English:** Generating the stack plot. `stackplot` creates the plot. `legend` adds a legend to the upper right. The title and labels for the axes are set, and a grid is added for better readability.
**Hindi:** Stack plot generate karna. `stackplot` plot banata hai. `legend` upar-dayen kone mein ek legend jodta hai. Axes ke liye title aur labels set kiye gaye hain, aur behtar readability ke liye ek grid joda gaya hai.

```python
plt.figure(figsize=(6,6))
plt.stackplot(days,stydy,tv,playing,gym,labels=label,colors=colors,alpha = 1.0)
plt.legend(loc='upper right')
plt.title('My activity')
plt.xlabel('activity',color = 'red')
plt.ylabel('hour',color = 'black')
plt.grid()
```
