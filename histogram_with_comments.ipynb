# Histograms using Matplotlib
**English:** This notebook shows how to create histograms using the Matplotlib library.
**Hindi:** Yeh notebook Matplotlib library ka upyog karke histogram banane ka tarika dikhata hai.

**English:** Importing the necessary libraries, `matplotlib.pyplot` for plotting and `numpy` for numerical operations.
**Hindi:** Zaroori libraries import karna, `matplotlib.pyplot` plotting ke liye aur `numpy` numerical operations ke liye.

```python
import matplotlib.pyplot as plt
import numpy as np
```

**English:** Creating a sample dataset using NumPy's `arange` function, which will be used to plot the histogram.
**Hindi:** NumPy ke `arange` function ka upyog karke ek sample dataset banana, jiska upyog histogram plot karne ke liye kiya jayega.

```python
data = np.arange(1,100)
# data
```

**English:** Generating the histogram. `bins=5` divides the data into 5 intervals. `edgecolor='red'` sets the color of the bin edges. The title and labels for the x and y axes are also set.
**Hindi:** Histogram generate karna. `bins=5` data ko 5 intervals mein divide karta hai. `edgecolor='red'` bin ke kinaron ka rang set karta hai. X aur y axis ke liye title aur labels bhi set kiye gaye hain.

```python
plt.hist(data,bins=5,edgecolor='red')
# plt.axhline(np.mean(data),color = '#D1D59C',linestyle='--',label='mean')
plt.title('youtube video views')
plt.xlabel('views')
plt.ylabel('frequency')
```
